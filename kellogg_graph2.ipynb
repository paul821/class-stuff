```python
# prompt: create a line graph of: Cybertruck Base: 39900 Nov 2021, 60990 Nov 2023
# Cybertruck AWD: 49990 Nov 2021, 79990 Nov 2023, 99990 Aug 2024, 79990 Oct 2024
# Cybertruck Most Expensive: 60990 Nov 2021, 99990 Nov 2023, 119990 Aug 2024, 99990 Oct 2024
# Ford F-150 Lightning Base 39974 May 2021, 46974 Aug 2022, 51974 Oct 2022, 55974 Dec 2022, 59974 Mar 2023, 59974 Oct 2024
# Ford F-150 Lightning Most Expensive 92000 May 2021, 97000 Aug 2022, 97000 Oct 2022, 97000 Dec 2022, 99969 Mar 2023, 99969 Oct 2024
# use dark green color scheme for ford and dark red color scheme for cybertruck. remove grids. put legend at the top left

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Data
cybertruck_base_data = {
    datetime(2021, 11, 1): 39900,
    datetime(2023, 11, 1): 60990
}
cybertruck_awd_data = {
    datetime(2021, 11, 1): 49990,
    datetime(2023, 11, 1): 79990,
    datetime(2024, 8, 1): 99990,
    datetime(2024, 10, 1): 79990
}
cybertruck_expensive_data = {
    datetime(2021, 11, 1): 60990,
    datetime(2023, 11, 1): 99990,
    datetime(2024, 8, 1): 119990,
    datetime(2024, 10, 1): 99990
}
ford_base_data = {
    datetime(2021, 5, 1): 39974,
    datetime(2022, 8, 1): 46974,
    datetime(2022, 10, 1): 51974,
    datetime(2022, 12, 1): 55974,
    datetime(2023, 3, 1): 59974,
    datetime(2024, 1, 3): 59974
    datetime(2024, 10, 1): 59974
}
ford_expensive_data = {
    datetime(2021, 5, 1): 92000,
    datetime(2022, 8, 1): 97000,
    datetime(2022, 10, 1): 97000,
    datetime(2022, 12, 1): 97000,
    datetime(2023, 3, 1): 99969,
    datetime(2024, 10, 1): 99969
}


# Plotting
plt.figure(figsize=(12, 6))

plt.plot(list(cybertruck_base_data.keys()), list(cybertruck_base_data.values()), marker='o', linestyle='-', color='darkred', label='Cybertruck Base')
plt.plot(list(cybertruck_awd_data.keys()), list(cybertruck_awd_data.values()), marker='o', linestyle='-', color='firebrick', label='Cybertruck AWD')
plt.plot(list(cybertruck_expensive_data.keys()), list(cybertruck_expensive_data.values()), marker='o', linestyle='-', color='red', label='Cybertruck Most Expensive')
plt.plot(list(ford_base_data.keys()), list(ford_base_data.values()), marker='o', linestyle='-', color='darkgreen', label='Ford F-150 Lightning Base')
plt.plot(list(ford_expensive_data.keys()), list(ford_expensive_data.values()), marker='o', linestyle='-', color='forestgreen', label='Ford F-150 Lightning Most Expensive')

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("F-150 Lightning and Cybertruck Pricing Changes until Oct 2024")

# Format x-axis as dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())

plt.grid(False) # remove grid

plt.legend(loc='upper left') # legend location
plt.gcf().autofmt_xdate() # rotate date labels
plt.show()
```
